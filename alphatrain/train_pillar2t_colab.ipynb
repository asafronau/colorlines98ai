# AlphaTrain Pillar 2T: The Topology Pivot

**THE FUNDAMENTAL CHANGE:** Value target switched from density-discounted TD returns
(SNR=0.03, IQR=4.4) to **sqrt(remaining_turns)** (IQR=29.1, 6.6x more contrast).

- Dying board (20 turns left): V = 4.5
- Tipping point (200 turns left): V = 14.1
- Healthy board (3000 turns left): V = 54.8
- Gap between "fine" and "dying": **40 points** (was 0.2)

Same V6 data (1,654 games, 5.4M states), same architecture.
max_score=85, 128 bins (0.67/bin), val_weight=0.1, rank_weight=1.0.
Warm start from 2R epoch 3.
