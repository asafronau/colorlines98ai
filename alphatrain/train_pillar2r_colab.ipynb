# AlphaTrain Pillar 2R: High-SNR Value Head Fix

**Diagnosis:** Value head SNR = 0.03 on general boards. Afterstate ranking gap=22 (already
solved), but board-level gap=0.4 (broken). MCTS evaluates boards, not afterstates.

**Root cause:** val_weight=0.01 gave value CE only 1% of gradient. Backbone never learned
value-discriminating features for general boards. Ranking on afterstates doesn't transfer.

**Fix (v2):**
- **val_weight=0.1** (10x increase) — force backbone to learn board health features
- Hard ranking pairs (top-1 vs top-5 MCTS), margin_target=10
- rank_weight=1.0

Expected loss balance: pol≈2.1, val_contribution≈0.4, rank≈1.2 → total≈3.7.
Value is now ~11% of gradient (was 1%). Policy still dominant.

Everything else unchanged: same v5 density data, γ=0.98, 128 bins, 50% endgame,
warm start from 2P ep6, 15 epochs.
